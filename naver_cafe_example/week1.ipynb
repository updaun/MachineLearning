## 2유형 고객의 신상정보 데이터를 통한 회사 서비스 이탈 예측 (종속변수 : Exited)

```python
trainData  = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/churn/train.csv'
testData  = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/churn/test.csv'
subData  = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/churn/submission.csv'

import pandas as pd

train = pd.read_csv(trainData)
test = pd.read_csv(testData)
submission = pd.read_csv(subData)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier


def preprocessing(df, target='train', encoderLst=None):

    drop_col = ['RowNumber','CustomerId','Surname']
    df = df.drop(drop_col, axis=1)

    if target == 'train':
        labelEncoder1 = LabelEncoder()
        labelEncoder2 = LabelEncoder()

        df['Geography'] = labelEncoder1.fit_transform(df['Geography'])
        df['Gender'] = labelEncoder2.fit_transform(df['Gender'])
        encoderLst = [labelEncoder1, labelEncoder2]
        return df, encoderLst
    
    elif target == 'test' and encoderLst != None:
        labelEncoder1 = encoderLst[0]
        labelEncoder2 = encoderLst[1]

        df['Geography'] = labelEncoder1.transform(df['Geography'])
        df['Gender'] = labelEncoder2.transform(df['Gender'])
        return df
    
    return None


preprocessing_df, encoderLst = preprocessing(train)


from sklearn.model_selection import train_test_split

X = preprocessing_df.drop('Exited', axis=1)
y = preprocessing_df['Exited']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify=y)

rfmodel = RandomForestClassifier(random_state=42)

rfmodel.fit(X_train, y_train)

y_pred = rfmodel.predict(X_test)

from sklearn.metrics import f1_score

print(f1_score(y_test, y_pred))

```

```python
preprocessing_test_df = preprocessing(test, 'test', encoderLst)

test_pred = rfmodel.predict(preprocessing_test_df)

print(test_pred)
# submission[0] = test_pred
# submission.to_csv('submission_model.csv', index=False)
```

### 유방암 발생여부 예측

```python
trainData  = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/cancer/train.csv'
testData  = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/cancer/test.csv'
subData  = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/cancer/submission.csv'
```

```python
import pandas as pd

train = pd.read_csv(trainData)
test = pd.read_csv(testData)
submission = pd.read_csv(subData)


from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier


def preprocessing(df, target='train', encoderLst=None):

    drop_col = ['id']
    df = df.drop(drop_col, axis=1)

    if target == 'train':
        scaler = StandardScaler()
        df = scaler.fit_transform(df)

        encoderLst = [scaler]
        return df, encoderLst

    elif target == 'test' and encoderLst != None:
        scaler = encoderLst[0]

        df = scaler.fit_transform(df)

        return df
    
    return None


X = train.drop(['diagnosis'], axis=1)
y = train['diagnosis'].map(lambda x:1 if x=='B' else 0)


preprocessingdf, encoderLst = preprocessing(X)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(preprocessingdf, y, test_size=0.33, random_state=42)

rf = RandomForestClassifier()

rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

from sklearn.metrics import f1_score

f1_score(y_test, y_pred)
```

```python
preprocessing_test_df = preprocessing(test, 'test', encoderLst)

test_pred = rf.predict(preprocessing_test_df)

print(test_pred)
# submission[0] = test_pred
# submission.to_csv('submission_model.csv', index=False)
```
