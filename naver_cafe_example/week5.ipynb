## 2유형

```python
Main_name = 'MedicalCost'

trainData  = f'https://raw.githubusercontent.com/Datamanim/datarepo/main/{Main_name}/train.csv'
testData  = f'https://raw.githubusercontent.com/Datamanim/datarepo/main/{Main_name}/test.csv'
subData  = f'https://raw.githubusercontent.com/Datamanim/datarepo/main/{Main_name}/submission.csv'


#데이터로드
import pandas as pd

train = pd.read_csv(trainData)
test = pd.read_csv(testData)
submission = pd.read_csv(subData)

target_col = 'charges'

#랜덤포레스트 분류기 
from sklearn.ensemble import RandomForestRegressor


#훈련데이터 분리
from sklearn.model_selection import train_test_split

X = train.drop(target_col,axis=1)
X = pd.get_dummies(X)
y = train[target_col]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

#모델학습 및 평가
rf = RandomForestRegressor(random_state=42)

rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)

from sklearn.metrics import r2_score

print(r2_score(y_test, y_pred))

##0.825


#제출 파일생성을 위해 test데이터에 대해서 전처리 진행
preprocessing_test_df  = pd.get_dummies(test)[X.columns]

test_pred = rf.predict(preprocessing_test_df)
print(test_pred)
```

## 2유형

```python
Main_name = 'kingcountyprice'

trainData  = f'https://raw.githubusercontent.com/Datamanim/datarepo/main/{Main_name}/train.csv'
testData  = f'https://raw.githubusercontent.com/Datamanim/datarepo/main/{Main_name}/test.csv'
subData  = f'https://raw.githubusercontent.com/Datamanim/datarepo/main/{Main_name}/submission.csv'


#데이터로드
import pandas as pd

train = pd.read_csv(trainData)
test = pd.read_csv(testData)
submission = pd.read_csv(subData)

train.head()
```

```python
train.isnull().sum()
```

```python
train.info()
```

```python
drop_col = ['id', 'date','zipcode']

target_col = 'price'

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X = train.drop([target_col] + drop_col, axis = 1)
X = pd.get_dummies(X)
y = train['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

rf = RandomForestRegressor()
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```

```python
test = test.drop(drop_col, axis = 1)
test = pd.get_dummies(test)

pred_test = rf.predict(test)

print(pred_test)
```
