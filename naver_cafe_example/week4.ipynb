## 2유형

```python
trainData  = f'https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv'
testData  = f'https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/test.csv'
subData  = f'https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/submission.csv'

target_col = 'satisfaction'
#데이터로드
import pandas as pd

train = pd.read_csv(trainData)
test = pd.read_csv(testData)
submission = pd.read_csv(subData)

train.head()
```

```python
#랜덤포레스트 분류기 
from sklearn.ensemble import RandomForestClassifier


#훈련데이터 분리
from sklearn.model_selection import train_test_split

drop_col = ['Arrival Delay in Minutes','id']

X = train.drop([target_col]+drop_col, axis = 1)
y = train[target_col]

X = pd.get_dummies(X)
y = y.map(lambda x : 1 if x=='satisfied' else 0)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify=y)

rf = RandomForestClassifier(random_state=42)

rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

from sklearn.metrics import f1_score

f1_score(y_test, y_pred)
```

```python
test = test.drop(drop_col, axis=1)
test = pd.get_dummies(test)

preprocessing_test_df = test[X.columns]

test_pred = rf.predict(preprocessing_test_df)

print(test_pred)
```

## 2유형

```python
trainData  = f'https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/train.csv'
testData  = f'https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/test.csv'
subData  = f'https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/submission.csv'


#데이터로드
import pandas as pd

train = pd.read_csv(trainData)
test = pd.read_csv(testData)
submission = pd.read_csv(subData)

train.head()
```

```python
train.isnull().sum()
```

```python
#랜덤포레스트 분류기 
from sklearn.ensemble import RandomForestClassifier


#훈련데이터 분리
from sklearn.model_selection import train_test_split

target_col ='Potability'
```

```python
train = train.fillna(train.median())
test = test.fillna(test.median())
```

```python
X = train.drop(target_col, axis=1)
y = train[target_col]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify=y)

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)


from sklearn.metrics import f1_score

f1_score(y_test, y_pred)

```

```python
test_pred = rf.predict(test)
print(test_pred)
```
