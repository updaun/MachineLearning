```python
# 소수점 나이 찾기
import pandas as pd

df = pd.read_csv("dataset/basic1.csv")
df
```

```python
import numpy as np

# 소수점 찾기
df = df[df['age']-np.floor(df['age']) != 0]
df
```

```python
# 올림
m_ceil = np.ceil(df['age']).mean()

# 내림
m_floor = np.floor(df['age']).mean()

# 버림
m_trunc = np.trunc(df['age']).mean()

m_ceil, m_floor, m_trunc
```

```python
print(m_ceil + m_floor + m_trunc)
```
